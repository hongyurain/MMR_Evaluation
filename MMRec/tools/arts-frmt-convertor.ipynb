## U-I graph

```python
import pandas as pd
```

```python
df = pd.read_csv('./arts.inter')
df.shape
```

```python
df[:5]
```

```python
print('# of items: ', len(df['item_id:token'].unique()))
print('# of users: ', len(df['user_id:token'].unique()))
```

```python
splitting = [0.7, 0.1, 0.2]
ts_id, x_label = 'timestamp:float', 'x_label'
rslt_dir = './'

tot_ratio = sum(splitting)
# remove 0.0 in ratios
ratios = [i for i in splitting if i > .0]
ratios = [_ / tot_ratio for _ in ratios]
split_ratios = np.cumsum(ratios)[:-1]
split_timestamps = list(np.quantile(df[ts_id], split_ratios))
# get df training dataset unique users/items
df_train = df.loc[df[ts_id] < split_timestamps[0]].copy()
df_val = df.loc[(split_timestamps[0] <= df[ts_id]) & (df[ts_id] < split_timestamps[1])].copy()
df_test = df.loc[(split_timestamps[1] <= df[ts_id])].copy()

df_train[x_label] = 0
df_val[x_label] = 1
df_test[x_label] = 2
temp_df = pd.concat([df_train, df_val, df_test])
temp_df[:5]
```

```python
temp_df.columns = ['itemID', 'userID', 'timestamp', 'x_label']
ret_df = temp_df[['userID', 'itemID', 'timestamp', 'x_label']]
ret_df[:5]
```

```python

temp_df.to_csv(os.path.join(rslt_dir, 'arts-labelled.inter'), sep='\t', index=False)
# df
print('done!')
```

## Features

```python
import pandas as pd
image_feat = pd.read_csv('arts.image')
image_feat.columns
```

```python
image_feat[:2]
```

```python
image_feat.sort_values(by=['iid_img:token'], inplace=True)
image_feat[-2:]
```

```python
df_feat = image_feat['img_emb:float_seq']
ret_df = df_feat.map(lambda x: list(eval(x)))
ret_df[0]
```

```python
ret_ls = ret_df.values.tolist()
ret_ls[0]
```

```python
np.save('art-img-feat.npy', np.array(ret_ls, dtype=np.float))
print('image feat done!')
```

```python
text_feat = pd.read_csv('arts.text')
text_feat.sort_values(by=['iid_txt:token'], inplace=True)
df_feat_t = text_feat['txt_emb:float_seq']
ret_df_t = df_feat_t.map(lambda x: list(eval(x)))

ret_ls_t = ret_df_t.values.tolist()
np.save('art-text-feat.npy',  np.array(ret_ls_t, dtype=np.float))
print('text feat done!')
```
